## **1. Data Exploration**

```python
train_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
df = pd.read_csv(train_url) #training set

df.head()
```

```python
df.tail()
```

```python
df.describe()
```

```python
test_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/test.csv"

df_test = pd.read_csv(test_url) #test set

df_test.head()
```

```python
df_test.describe()
```

### missing value

```python
# Check the number of instance which contains N/A data
nan_in_df = df.isna().any(axis=1).sum()
print('Number of NaN values in train set: ' + str(nan_in_df))
```

```python
train_drop = df.dropna(axis = 0)
nan_in_df = train_drop.isna().any(axis=1).sum()
print('Number of NaN values in train set: ' + str(nan_in_df))
```

```python
# Check the number of instance which contains N/A data
nan_in_df = df_test.isna().any(axis=1).sum()
print('Number of NaN values in test set: ' + str(nan_in_df))
```

```python
test_drop = df_test.dropna(axis = 0)
nan_in_df = test_drop.isna().any(axis=1).sum()
print('Number of NaN values in test set: ' + str(nan_in_df))
```

```python
df.describe()
```

```python
mode_age = df["Age"].mode()[0]  # Extracting the mode value
df["Age"] = df["Age"].fillna(mode_age)
df["Age"] = df["Age"].fillna(df["Age"].mode()[0])
print(mode_age)
df.tail()
```

```python
df.loc[df["Embarked"] == "S", "Embarked"] = 0
df.loc[df["Embarked"] == "C", "Embarked"] = 1
df.loc[df["Embarked"] == "Q", "Embarked"] = 2
df.head()
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
df.tail()
```

```python
df.loc[df["Sex"] == "female", "Sex"] = 0
df.loc[df["Sex"] == "male", "Sex"] = 1
df.head()
```

```python
df["Sex"] = df["Sex"].fillna(df["Sex"].mode()[0])
df.tail()
```

```python
ratio = 0.75
 
total_rows = df.shape[0]
train_size = int(total_rows*ratio)
 
# Split data into test and train
train = df[0:train_size]
test = df[train_size:]

print("Train dataframe")
print(train)
 
# print test set
print("Test dataframe")
print(test)
```

```python
x_train = np.array(train[["Pclass","Sex","Age","Embarked"]].values, dtype = float)
# x_train = np.array(train[["Sex","Age"]].values, dtype = float)
print(x_train)
print(x_train.shape)
```

```python
y_train = np.array(train[["Survived"]].values, dtype = int)
print(y_train)
print(y_train.shape)
```

```python
x_test = np.array(test[["Pclass","Sex","Age","Embarked"]].values, dtype = float)
print(x_test)
print(x_test.shape)
```

```python
y_test = np.array(test[["Survived"]].values, dtype = int)
print(y_test)
print(y_test.shape)
```

```python
model = Regression('logistic',0.0012,50000)
model.fit(x_train, y_train)

print(model.MSE_history)
```

```python
plt.plot(range(1, model.num_iterations + 1), model.MSE_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Cost Function')
plt.grid(True, linestyle='--', linewidth=0.4, which='both')
plt.minorticks_on()
plt.grid(True, linestyle='--', linewidth=0.4, which='minor')
```

```python
train_predictions,list_predict,list_prob = model.predict(x_train,0.45)
# test_predictions,list_predict1,list_prob1 = model.predict(x_test)

Accuracy_model = Accuracy(y_train, list_predict)

# print("model :",list_prob)
print("train_predictions :",train_predictions)
print("Accuracy :", Accuracy_model.accuracy())
print("Precision :", Accuracy_model.precision())
print("Recall :", Accuracy_model.recall())
print("f1_score :", Accuracy_model.f1_score())
```

```python
accuracy = []
precision = []
recall = []
f1_score = []

thresholds = np.arange(0, 1.01, 0.01)  # Vary threshold from 0 to 1 in steps of 0.01

for threshold in thresholds:
    train_predictions, list_predict, list_prob = model.predict(x_train, threshold)
    Accuracy_model = Accuracy(y_train, list_predict)
    accuracy.append(round(Accuracy_model.accuracy(), 4))
    precision.append(round(Accuracy_model.precision(), 4))
    recall.append(round(Accuracy_model.recall(), 4))
    f1_score.append(round(Accuracy_model.f1_score(), 4))
    

plt.plot(thresholds, accuracy, label="accuracy") 
plt.plot(thresholds, precision, label="precision")
plt.plot(thresholds, recall, label="recall")
plt.plot(thresholds, f1_score, label="f1_score")
plt.xlabel('Threshold')
plt.ylabel('Accuracy')
plt.grid(True, linestyle='--', linewidth=0.4, which='both')
plt.minorticks_on()
plt.grid(True, linestyle='--', linewidth=0.4, which='minor')
plt.title('Accuracy vs. Threshold')
leg = plt.legend(loc='lower left', fontsize=7, ncol=4)
plt.show()
```

```python
x_train_copy = x_train.copy()
high_column = x_train_copy[:, 2] * x_train_copy[:, 0]
x_train_copy[:, 2] = np.round(high_column, 2)
print(x_train_copy)

# Instantiate and train the model with higher-order features
model_higher_order = Regression('logistic',0.0055,50000)
model_higher_order.fit(x_train_copy, y_train)

# Predictions on training set with higher-order features
_, train_predictions_higher_order, _ = model_higher_order.predict(x_train_copy,0.1)

# Calculate accuracy, precision, recall, and F1-score on the training set with higher-order features
accuracy_train_higher_order = Accuracy(y_train, train_predictions_higher_order).accuracy()
precision_train_higher_order = Accuracy(y_train, train_predictions_higher_order).precision()
recall_train_higher_order = Accuracy(y_train, train_predictions_higher_order).recall()
f1_score_train_higher_order = Accuracy(y_train, train_predictions_higher_order).f1_score()

print("Training Set Metrics with Higher-Order Features:")
print("Accuracy:", accuracy_train_higher_order)
print("Precision:", precision_train_higher_order)
print("Recall:", recall_train_higher_order)
print("F1-score:", f1_score_train_higher_order)
```

```python
accuracy_high = []
precision_high = []
recall_high = []
f1_score_high = []

thresholds = np.arange(0, 1.01, 0.01)  # Vary threshold from 0 to 1 in steps of 0.01

for threshold in thresholds:
    _, train_predictions_higher_order, _ = model_higher_order.predict(x_train_copy, threshold)
    Accuracy_model_high = Accuracy(y_train, train_predictions_higher_order)
    accuracy_high.append(round(Accuracy_model_high.accuracy(), 4))
    precision_high.append(round(Accuracy_model_high.precision(), 4))
    recall_high.append(round(Accuracy_model_high.recall(), 4))
    f1_score_high.append(round(Accuracy_model_high.f1_score(), 4))

plt.plot(thresholds, accuracy_high, label="accuracy") 
plt.plot(thresholds, precision_high, label="precision")
plt.plot(thresholds, recall_high, label="recall")
plt.plot(thresholds, f1_score_high, label="f1_score")
plt.xlabel('Threshold')
plt.ylabel('Accuracy')
plt.grid(True, linestyle='--', linewidth=0.4, which='both')
plt.minorticks_on()
plt.grid(True, linestyle='--', linewidth=0.4, which='minor')
plt.title('Accuracy vs. Threshold')
leg = plt.legend(loc='lower left', fontsize=7, ncol=4)
plt.show()
```
